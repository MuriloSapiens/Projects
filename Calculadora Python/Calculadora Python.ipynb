```python
print("\n************Calculadora*************")

def add(x,y):
  return x + y
def subtract(x,y):
  return x - y
def multiply(x,y):
  return x * y
def divisao(x,y):
  return x / y

print("\nEscolha uma operação\n")
print("1-Adição")
print("2-Subtração")
print("3-Multiplicão")
print("4-Divisão")

operaçao = input("\nOperação:")
numero1 = int(input("\nPrimeiro Número:"))
numero2 = int(input("\nSegundo Número:"))

if operaçao == '1':
  print("\n")
  print(numero1,'+',numero2,'=',add(numero1,numero2))

elif operaçao == '2':
  print("\n")
  print(numero1,'-',numero2,'=',subtract(numero1,numero2))

elif operaçao == '3':
  print("\n")
  print(numero1,'*',numero2,'=',multiply(numero1,numero2))

elif operaçao == '4':
  print("\n")
  print(numero1,'/',numero2,'=',divide(numero1,numero2))

else:
  print('\nO dado fornecido para operações está invalido!!!') 
```
